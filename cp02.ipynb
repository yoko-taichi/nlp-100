### p12.1列目をcol1.txtに、2列目をcol2.txtに保存

```python
with open("popular-names.txt", "r") as f, open("col1.txt", "w") as fc1, open("col2.txt", "w") as fc2:
    lines = f.readlines()
    count = len(lines)
    for i, line in enumerate(lines):
        name, color, _, _ = line.split("\t")
        if count == i:
            fc1.write(name)
            fc2.write(color)
        else:
            fc1.write(name + '\n')
            fc2.write(color + '\n')
```

### p16.ファイルをN分割する

```python
import numpy as np
N = int(input())
with open('popular-names.txt', 'r') as f:
    lines = f.readlines()
    count = len(lines)
    num_list = range(count)
    div_list = np.array_split(num_list, N)
    for i, div in enumerate(div_list, 1):
        f = open('{}.txt'.format(str(i).zfill(3)), 'w')
        for j in div:
            f.write(lines[j])
        f.close()
            
        
```

### p17.1列目の文字列の異なり
1列目の文字列の種類(異なる文字列の集合)を求めよ。確認にはcut,sort,uniqコマンドを用いよ。

```python
with open('popular-names.txt', 'r') as f:
    lines = f.readlines()
    name = sorted(set(map(lambda x: x.split('\t')[0], lines)))
print(name)
```

```python
! cut -f 1 "popular-names.txt"|sort|uniq
```

### p18.各行を3コラム目の数値の降順にソート

```python
with open("popular-names.txt", "r") as f:
    lines = f.readlines()
    lines = map(lambda x: x.replace("\n",""), lines)
    temp = list(map(lambda x: x.split('\t'), lines))
    pnum = sorted(temp, key=lambda x: int(x[2]), reverse=True)
print(pnum)
```

```python
! sort -n -r -k 3,3 -t$" " "popular-names.txt"
```

### p19.各行の1コラム目の文字列の出現頻度を求め、出現頻度の高い順に並べる

```python
import collections
with open("popular-names.txt", "r") as f:
    lines = f.readlines()
    temp = list(map(lambda x: x.split('\t'), lines))
    c = collections.Counter(map(lambda x: x[0], temp))
    c2 = c.most_common()

print(c2)
```

```python
! cut  '\t' -f 1 "popular-names.txt"|sort|uniq -c|sort -r -n
```
